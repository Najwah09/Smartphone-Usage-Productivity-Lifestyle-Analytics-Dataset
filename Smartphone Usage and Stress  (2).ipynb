# Correction DataType

Based on the df.info() output, the 'Gender', 'Occupation', and 'Device_Type' columns are identified as categorical variables currently stored as generic object types. Converting them to the category dtype will improve memory efficiency and ensure proper handling of these variables in subsequent analyses.

```python
df.info()
```

```python
df['Gender'] = df['Gender'].astype('category')
df['Occupation'] = df['Occupation'].astype('category')
df['Device_Type'] = df['Device_Type'].astype('category')

print("Data types after conversion:")
df.info()
```

# Explore and Create Derived Columns
Analyze the existing columns to identify opportunities for creating new, derived features that could provide additional insights or simplify analysis, such as total screen time.

- creating a new column 'Total_Screen_Time' by summing specified existing columns and then verifying its creation and displaying its descriptive statistics. This single code block will perform all these actions sequentially.

```python
df["Total_Screen_Time"]=df["Daily_Phone_Hours"]+df["Social_Media_Hours"]+ df['Weekend_Screen_Time_Hours']
print("With Data the 'Total Screen Time' column in First 5 rows:")
print(df.head())
print("\nDescriptive statistics for 'Total_Screen_Time':")
print(df['Total_Screen_Time'].describe())
```

# Filtering or aggregating data
Perform preliminary filtering or aggregation of the data based on relevant columns (e.g., 'Occupation', 'Device_Type') to understand distributions or group-level statistics.

- To understand the distribution and group-level statistics for 'Occupation', we group the DataFrame by 'Occupation' and calculate the mean of 'Daily_Phone_Hours', 'Work_Productivity_Score', and 'Total_Screen_Time'.

```python
occupation_stats = df.groupby('Occupation', observed=False)[['Daily_Phone_Hours', 'Work_Productivity_Score', 'Total_Screen_Time']].mean()
print("Mean 'Daily_Phone_Hours', 'Work_Productivity_Score', and 'Total_Screen_Time' by Occupation:")
print(occupation_stats)

print("\n---\n")

device_stats = df.groupby('Device_Type', observed=False)[['Daily_Phone_Hours', 'Work_Productivity_Score', 'Total_Screen_Time']].mean()
print("Mean 'Daily_Phone_Hours', 'Work_Productivity_Score', and 'Total_Screen_Time' by Device Type:")
print(device_stats)
```

# 7. Correlation HeatMap

```python
plt.figure(figsize=(10,8))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')

plt.title("Correlation Matrix of Numerical Variables")

plt.show()
```

Insight:- 
- A heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors. For a correlation matrix, this means that the strength and direction of the correlations are visually apparent through a color gradient.
- This Correlation Matrix heatmap shows how much different variables "move together." The numbers range from -1 to +1, where 1.0 (dark red) is a perfect positive relationship and 0 is no relationship at all.
  
1. The Only Strong Connections (0.4 to 0.7)
The only variables that actually relate to each other are the time-based ones. These are the "warm" colored boxes in the bottom right corner:
Total Screen Time vs. Weekend Screen Time (0.67): A strong positive correlation. If someone uses their phone a lot on weekends, their total weekly usage is naturally higher.
Total Screen Time vs. Daily Phone Hours (0.61): Another strong link, which makes logical sense.
Total Screen Time vs. Social Media Hours (0.41): A moderate link, suggesting social media is a big part of screen time, but not the only part.
2. Near-Zero Correlation (Everything Else)
Almost every other number in this table is close to 0.00 (appearing as light blue or neutral). This is a major finding because it shows no relationship between:
Age and Screen Time: Age has a correlation of only 0.0035 with Screen Time. This means in your dataset, an 18-year-old and a 60-year-old have virtually the same phone habits.
Productivity and Screen Time: The correlation is -0.0035. This proves that in this specific dataset, spending more time on your phone does not lower your productivity score.
Sleep and Stress: Interestingly, there is almost zero link here (0.0047), which is unusual for real-world data where less sleep usually equals more stress.
3. The Diagonal "1.0" Line
The dark red line of 1s running from top-left to bottom-right is simply each variable correlating with itself (e.g., Age vs. Age). You can ignore this for your analysis.

Summary for your report:
- The dataset shows that phone usage habits (Social Media, Weekend time, Daily hours) are interconnected, but these habits have no measurable impact on Age, Stress, Sleep, or Productivity.

# 10 . Pair Plot

-A pair plot is an excellent tool for visualizing the pairwise relationships between multiple numerical variables in a dataset. It generates a grid of scatter plots for each pair of variables and histograms or KDE plots for the distribution of each single variable. This helps in quickly identifying trends, correlations, and anomalies among key features related to screen time, productivity, sleep, and stress.

```python

selected_numerical_cols = [
    'Daily_Phone_Hours',
    'Social_Media_Hours',
    'Work_Productivity_Score',
    'Sleep_Hours',
    'Stress_Level',
    'Total_Screen_Time'
]

print("Generating pair plot for selected numerical variables...")
sns.pairplot(df[selected_numerical_cols])
plt.suptitle('Pair Plot of Key Numerical Variables', y=1.02) # Adjust suptitle to not overlap with axes
plt.show()
```


The pair plot is a powerful visualization tool that helps us understand both the individual distributions of several numerical variables and the relationships between each pair of those variables.

Here's how to interpret it:
Diagonal Plots (Histograms/KDEs):

These plots run from the top-left to the bottom-right of the grid.
They show the distribution of each individual variable (Daily_Phone_Hours, Social_Media_Hours, Work_Productivity_Score, Sleep_Hours, Stress_Level, Total_Screen_Time).
From these, you can see if a variable's values are normally distributed, skewed, bimodal, or have any other particular shape. For example, you can see the typical range of Sleep_Hours or whether Stress_Level is more concentrated at certain values.
Off-Diagonal Plots (Scatter Plots):

These are the plots that fill the rest of the grid.
Each scatter plot shows the relationship between two different variables.
For instance, if you look at the intersection of the Total_Screen_Time row and Work_Productivity_Score column, you'll see a scatter plot where Total_Screen_Time is on the x-axis and Work_Productivity_Score is on the y-axis
